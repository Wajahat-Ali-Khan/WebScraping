```python
from autoscraper import AutoScraper
```

```python
url = 'https://www.olx.com.pk/items/q-suzuki-150'
# We can add one or multiple candidates here.
# You can also put urls here to retrieve urls.
wanted_list = ["Orangi Town, Karachi",
               "super power leo heavy bike 200 Honda150 Suzuki150 unique better than",
               "Rs 185,000"]
#                "guaranteed price  SALE  offer front and back fender only Rs: 3000 offer includes 1: front fenderGS 150  SE 2021 plastic 2: back fender GS 150 SE  available in black and red colors we also have safeguard ,ladies footrest ,LEDs all accessories of GS 150 visit our store at m. a Jinnah road Aurangzeb market mumtaz traders timings Mon to sat 11 am - 8 pm delivery also available all over Pakistan charges 200"]
```

```python
scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
#Grouping For Sperating Title, Proce and Location

scraper.get_result_similar("https://www.olx.com.pk/items/q-suzuki-150",grouped=True)
```

```python
# Set and save the rules for our results

scraper.set_rule_aliases({'rule_c2j4': 'Locations', 'rule_9ku1': 'Title','rule_qtca':'Price'})
scraper.keep_rules(['rule_c2j4', 'rule_9ku1','rule_qtca'])
scraper.save('Olx Suzuki 150')
```

```python
scraper.load('Olx Suzuki 150')
```

```python
#Just Change Url for similar Result for another product
result=scraper.get_result_similar("https://www.olx.com.pk/items/q-oppo",group_by_alias=True)
```
